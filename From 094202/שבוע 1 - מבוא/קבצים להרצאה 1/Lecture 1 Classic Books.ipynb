# Classic Books

Here, we explore a bit two classic books: *Peter Pan* and *Little Women*. These and other old books are now in the public domain, and [Project Gutenberg](http://www.gutenberg.org/) is a website that publishes such books. 
We will read the books from the website.

For now, do not worry about the code and focus more on interpreting the outputs.

```python
# Loading some packages and adjusting some settings
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from urllib.request import urlopen 
import re
def read_url(url): 
    return re.sub('[ \n\r]+', ' ', urlopen(url).read().decode())
# sns.set(style='ticks',font_scale=1.2)
# sns.set_palette("deep")
```

First, we read the two books from the webiste and put them into a dataframe (more on this later), with each row including one chapter.  

```python
# Read and parse Peter Pan to its chapters, then print out the list of chapters
peter_pan_url = 'http://www.gutenberg.org/files/16/16-0.txt'
peter_pan_text = read_url(peter_pan_url)
peter_pan_chapters = peter_pan_text.split('Chapter ')[18:]
peter_pan_df = pd.DataFrame(peter_pan_chapters, columns =['Chapters'])
peter_pan_df # Jupyter prints to screen the output of the last line of code
```

We can see that each chapter begins with a chapter number and the chapter's title in upper case.

We can also print out the first chapter explicitly:

```python
# Print the first chapter from Peter Pan
peter_pan_chapters[0]
```

The same for Little Women:

```python
little_women_url = 'http://www.gutenberg.org/cache/epub/514/pg514.txt'
little_women_text = read_url(little_women_url)
little_women_chapters = little_women_text.split('CHAPTER ')[48:]
little_women_df = pd.DataFrame(little_women_chapters, columns =['Chapters'])
little_women_df
```

This book has more chapters than Peter Pan, and the chapter number is written in upper case words.

## Literary charecters

Now that we loaded the texts, we can quickly visualize how many times each charecter is mentioned in the book. **(üö®Spoliler Alert!üö®)**

#### Peter Pan

For Peter Pan, we will compare the mentions of Peter, Wendy, John, Michael, Mrs. Darling, and Captain Hook.

```python
# Let's plot the number of mentions of each character's name in the book

# First, count for each character the number of mentions in each chapter
peter = np.cumsum(np.char.count(peter_pan_chapters, 'Peter'))
wendy = np.cumsum(np.char.count(peter_pan_chapters, 'Wendy'))
john = np.cumsum(np.char.count(peter_pan_chapters, 'John'))
michael = np.cumsum(np.char.count(peter_pan_chapters, 'Michael'))
hook = np.cumsum(np.char.count(peter_pan_chapters, 'Hook'))
mrs_d = np.cumsum(np.char.count(peter_pan_chapters, 'Mrs. Darling'))

chapter = list(range(1,18))

# plot the counts
plt.figure(figsize=(8,8))
plt.plot(chapter, peter, label="Peter")
plt.plot(chapter, wendy, label="Wendy")
plt.plot(chapter, john, label="John")
plt.plot(chapter, michael, label="Michael")
plt.plot(chapter, hook, label="Captain Hook")
plt.plot(chapter, mrs_d, label="Mrs. Darling")
plt.legend()
plt.xlabel('Chapter number')
plt.title('Cumulative number of times each name appears');
```

In the plot above, the x-axis shows chapter numbers and the y-axis shows how many times each character has been mentioned up to and including that chapter.

## Printed characters

In some situations, the relationships between quantities allow us to make predictions. As an example of visualizing information derived from multiple sources, let us first use the computer to get some information that would be tedious to acquire by hand: We want to count the number of characters (as in letters or punctuation marks) and the number of periods in each chapter of both books.

```python
# Count the number of characters and the number of periods in each chapter 
# and add this information to the dataframe of chapters

# Peter Pan
peter_pan_df['number of characters'] = peter_pan_df['Chapters'].str.count('.')
peter_pan_df['number of periods'] = peter_pan_df['Chapters'].str.count('\.')
display(peter_pan_df)

# Little Women
little_women_df['number of characters'] = little_women_df['Chapters'].str.count('.')
little_women_df['number of periods'] = little_women_df['Chapters'].str.count('\.')
little_women_df
```

Not surprisingly, chapters with fewer characters tend to have fewer periods, in general: the shorter the chapter, the fewer sentences there tend to be, and vice versa. The relation is not entirely predictable, however, as sentences are of varying lengths and can involve other punctuation such as question marks. 

We can also see that the chapters of *Little Women* are in general longer than those of *Peter Pan*, with the excpetion of the last chapter. 
***
Let us see if these two simple variables ‚Äì the length and number of periods in each chapter ‚Äì can tell us anything more about the two books. One way to do this is to plot both sets of data on the same axes. 

In the plot below, there is a dot for each chapter in each book. Green dots correspond to *Peter Pan* and red dots to *Little Women*. The x-axis represents the number of periods and the y-axis represents the number of characters.

```python
plt.figure(figsize=(12,8))
plt.scatter(peter_pan_df['number of periods'], peter_pan_df['number of characters'], label="Peter Pan", color="green")
plt.scatter(little_women_df['number of periods'], little_women_df['number of characters'], label="Little Women", color='red')
plt.legend()
plt.xlabel('Number of periods (sentences?) in chapter')
plt.ylabel('Number of characters in chapter')
plt.grid(True)
```

The plot shows us that many of the chapters of *Little Women* are longer than those of *Peter Pan*, although the number of periods does not necessarily differs much between the two books. It seems that *Peter Pan* has generally shorter sentences than *Little Women*. 

Furthermore, notice how the red points are roughly clustered around a straight line, as are the green points. That it, there seems to be a linear relation between the number of sentneces and the number of characters, from which we can deduce the number of characters in an average sentence. By looking at the plot, it seems that in *Peter Pan* it is about 80-120 characters per sentence and in *Little Women* it is about 100-140 characters per sentence.
