```python
#Import libraries:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style as style
style.use('tableau-colorblind10')
import seaborn as sns
sns.set_palette("viridis")
from tqdm import tqdm
import warnings
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
df_Mroz = pd.read_csv('Mroz.csv')
df = df_Mroz.drop(columns=['Unnamed: 0']) 
df
```

```python
for col in df.columns.unique():
    if type(df[col] == 'object'):
        df.replace({col: {'yes': 1, 'no': 0}}, inplace=True)
        df[col] = df[col].astype(int)

df.head(10)
```

```python
def func(row):
    return row['child618'] + row['child6']

df['child018'] = df.apply(func, axis='columns')

mean_of_child018 = df.child018.mean()

def func2(row):
    if row['child018'] > mean_of_child018:
        return 1
    else:
        return 0

df['large_family'] = df.apply(func2, axis='columns')

def func3(row):
    if row['income'] < df.income.mean():
        return 1
    else:
        return 0

df['is_less_than_mean_income'] = df.apply(func3, axis='columns')

def func4(row):
    if row['is_less_than_mean_income'] == 1 and row['large_family'] == 1:
       return 1
    else:
        return 0

df['needs_help'] = df.apply(func4, axis='columns')
df
```

```python
df_trimmed = df[((df['hoursw'] == 0) & (df['wagew'] == 0)) |
                     ((df['hoursw'] > 0) & (df['wagew'] > 0)) |
                     ((df['hoursw'] == 0) & (df['wagew'] > 0)) | ((df['hoursh'] == 0) & (df['wageh'] == 0)) |
                     ((df['hoursh'] > 0) & (df['wageh'] > 0)) | ((df['hoursh'] == 0) & (df['wageh'] > 0))  ]

# מחשבים את מטריצת הקורלציות
corr_trimmed = df_trimmed.corr(method='pearson')
# ממירים את המטריצה למערך, ממיינים אותו לפי ערכי הקורלציה לעמודת 'wagew'
corr_trimmed_sort = corr_trimmed.loc[corr_trimmed['needs_help'].abs().sort_values(ascending=False).index]

# מדפיסים את הקורלציות לעמודת 'wagew'
print(corr_trimmed_sort['needs_help'])

```

```python
# compute correlation between each pair of variables in data frame
correlations = df_trimmed.corr()

# plot heat map
plt.figure(figsize=(15,15))
g=sns.heatmap(correlations,annot=True)
```

```python
X = df_trimmed[['ageh','hoursw','educh']] # features
Y = df_trimmed['needs_help'] # labels
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20)  
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

```python
pd.DataFrame(data = scaled_X_train , columns = X_train.columns).head()
```

```python

warnings.simplefilter(action='ignore', category=FutureWarning)
kNN = KNeighborsClassifier(n_neighbors=6) # K=5
kNN.fit(X_train, Y_train)
Y_pred = kNN.predict(X_test)
```

```python
total_observations = len(Y_test)
total_matches = 0
for pred_value, real_value in zip(Y_pred, Y_test):
    if pred_value == real_value:
        total_matches += 1
        
print (f"The accuracy of the classifier is: {total_matches/total_observations}")
```
